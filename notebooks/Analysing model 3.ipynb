```python
def sq(x):
    return x*x
sq = np.vectorize(sq)
```

```python
math.sqrt(np.mean(sq(diff)))
```

```python
def gen_samples(N):
    samples = []
    for k in range(N):
        X_train_sampled = sampling_augmentation(X_train, 5)
        pred = model.predict(X_train_sampled,batch_size=64,verbose=True)
        pred = [x[0] for x in pred]
        diff = y_train-pred
        samples.append(math.sqrt(np.mean(sq(diff))))
    return samples
```

```python
def boosting(N):
    all_preds = []
    samples = []
    for k in range(N):
        X_train_sampled = sampling_augmentation(X_train, 5)
        pred = model.predict(X_train_sampled,batch_size=64,verbose=True)
        pred = [x[0] for x in pred]
        all_preds.append(pred)
        diff = y_train-pred
        samples.append(math.sqrt(np.mean(sq(diff))))
    
    return all_preds,samples
```

```python
all_preds, samples = boosting(50)
```

```python
ultimate_pred = np.mean(all_preds,axis=0)

diff = ultimate_pred - y_train

math.sqrt(np.mean(sq(diff)))

print samples

for i in range(len(all_preds)):
    ultimate_pred = np.mean(all_preds[:i+1],axis=0)
    diff = ultimate_pred - y_train
    print math.sqrt(np.mean(sq(diff)))
```
