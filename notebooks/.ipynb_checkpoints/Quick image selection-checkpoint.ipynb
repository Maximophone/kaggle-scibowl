In this script I try to reorganise and filter the data in a better way. This is a first draft.

First, for each image, we compute the average pixel value. The max and min of these will correspond to the systole and diastole(roughly...)

Then we need to select the 6 best consecutive slices. Why 6? because every patient in the dataset but 3 have at least 8 slices. Every patient in the dataset but 4 have at least 6 consecutive slices.
We will ignore the outliers for now. 

To select the best slices we look for the 6 consecutive slices that have the highest average of max-min pixels, ie rougly the most variation in the images between systole and diastole



Change of plans: we select the best 8 consecutive slices. From these we will sample randomly 5 slices. This should increase sample size

```python
import matplotlib.pyplot as plt

from skimage.restoration import denoise_tv_chambolle

from datatools import Data
from framework.config import LOCS

%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize'] = (15.0, 10.0)

import numpy as np
import cPickle as pickle
```

```python
train = Data(LOCS.test_dir)
```

```python
file_name = 'selector_test_8.p'
```

```python
N_cons = 8
```

```python
def crop(img):
    short_edge = min(img.shape[:2])
    yy = int((img.shape[0] - short_edge) / 2)
    xx = int((img.shape[1] - short_edge) / 2)
    crop_img = img[yy: yy + short_edge, xx: xx + short_edge]
    x = crop_img.shape[0]
    reduction = x/4
    crop_img = crop_img[reduction:x-reduction,reduction:x-reduction]
    
    return crop_img
```

```python
def get_vstats(images):
    vstats = []
    for im in images:
        vstats.append(im.mean())
    return vstats
```

```python
def find_best_consecutive(diff_list,n=6):
    max_score = 0
    best_i = 0
    for i in range(len(diff_list)-n+1):
        part_list = diff_list[i:i+n]
        #maybe tweak this score function...
        score = sum(part_list)/(np.std(part_list)+1)
        if score>max_score: 
            max_score = score
            best_i = i
    return (best_i,best_i+n)
        
```

```python
patients = train.samples.keys()
```

```python
results = {}

for patient in patients:
    train._cache.reset()
    print 'processing patient %s'%patient
    sax_slices = [sl for sl in train.samples[patient].keys() if 'sax' in sl]
    sax_slices.sort(key=lambda x:int(x.split('_')[1]))
    if len(sax_slices)<8: continue
    all_vstats = []
    vdiffs = []
    vmaxmins = {}
    for sl in sax_slices:
        images = train.samples[patient][sl]
        images.revert()
        images.apply(crop)
        vstats = get_vstats(images.imagelist)
        im_vmax = vstats.index(max(vstats))+1
        im_vmin = vstats.index(min(vstats))+1
        vdiff = max(vstats) - min(vstats)
        vdiffs.append(vdiff)
        vmaxmins[sl]=(im_vmax,im_vmin)
        all_vstats.append(vstats)
    start_slice,end_slice = find_best_consecutive(vdiffs,n=N_cons)
    slices_to_keep = sax_slices[start_slice:end_slice]
    results[patient] = {sl:{'vmax':vmaxmins[sl][0],'vmin':vmaxmins[sl][1]} for sl in slices_to_keep}

print 'dumping results'
pickle.dump(results,open(file_name,'wb'))
print 'done'
```
