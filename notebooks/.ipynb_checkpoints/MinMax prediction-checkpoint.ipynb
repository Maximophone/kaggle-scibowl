```python
def crop(img):
    short_edge = min(img.shape[:2])
    yy = int((img.shape[0] - short_edge) / 2)
    xx = int((img.shape[1] - short_edge) / 2)
    crop_img = img[yy: yy + short_edge, xx: xx + short_edge]
    x = crop_img.shape[0]
    reduction = x/4
    crop_img = crop_img[reduction:x-reduction,reduction:x-reduction]
    
    return crop_img
```

```python
def preprocess(weight):
    def inner(img):
        return denoise_tv_chambolle(img,weight=weight)
    return inner
```

```python
def get_vstats(images):
    vstats = []
    for im in images:
        vstats.append(im.mean())
    return vstats
```

```python
def max_consecutive(l):
    curr_max = 1
    max_temp = 1
    prev = l[0]
    for x in l[1:]:
        if x==prev+1:
            max_temp += 1
            if max_temp>curr_max: curr_max = max_temp
        else:
            max_temp=1
        prev = x
    return curr_max
```

```python
def minmax(images):
    v_stats = []
    for im in images:
        im_crop = crop(im)
        v_stats.append(im_crop.mean())
    return v_stats.index(min(v_stats)), v_stats.index(max(v_stats))
```
